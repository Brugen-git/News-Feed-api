Importing all the required libraries

```python
import requests
import json
import os
import pandas as pd
import time
from selenium import webdriver
from bs4 import BeautifulSoup
```

Creating api class and defining the functions

```python
class API:
    
    def __init__ (self, keyword):
        self.keyword = keyword
        
    def json_display(self, obj):
        # Creating a string of JSON object
        with open('data.txt', 'w') as json_file:
            json.dump(obj, json_file)
        text = json.dumps(obj, sort_keys=True, indent=4)
        print(text)
        
    def create_df(self, obj):  
        # Creating a dataframe from nested objects
        FIELDS = ["source.id", "source.name", "author", "title", "description", "url", "urlToImage", "publishedAt", "content"]
        df = pd.json_normalize(obj['articles'])
        final_df = df[FIELDS]

        display(final_df.head(10))
        final_df.to_csv('data.csv', sep=',', index=False,header=True)
    
    def news_api(self):
        # Using the news api and searching the  keyword
        url = ('https://newsapi.org/v2/everything?'
       'q={keyword}&'
       'apiKey=4e70cabb80884db08524a28ac33cdc1d'.format(keyword = self.keyword))
        
        
        response = requests.get(url)
        if (response.status_code == 200):
            print('Successfuly able to call the API!')
            json_response = response.json()
            if(len(json_response['articles']) == 0):
                print('No news articles found for the keyword')
            else:
                
                # Print output in json
                self.json_display(json_response)     
                
                # Create a dataframe
                self.create_df(json_response)
                    
        else:
            print('Status code: ', response.status_code)
```

Entering keywords and calling the API

```python
keyword = input('Enter Keyword to be searched: ').lower()
w_api = API(keyword)
w_api.news_api()
```
